```python
import numpy as np
a_np = np.array([[1, 2, 3], [4, 5, 6]])
b_np = np.array([[1, 3, 5], [2, 4, 6]])

r1 = np.any(a_np > 3) 
r2 = np.all((a_np > 3)) 
r3 = a_np >= b_np
r4 = a_np[a_np >= b_np]
r5 = np.where(a_np > 2)
r6 = np.where(a_np > 2, a_np, 0)
r7 = np.where(a_np > 2, a_np*2, 1)
r8 = np.argmax(a_np)
r9 = np.argmax(a_np, axis=1)
r10 = np.argmax(a_np, axis=0)
```

```python
import numpy as np
a_np = np.array([[1, 2, 3], [4, 5, 6]])
b_np = np.array([[1, 3, 5], [2, 4, 6]])

r1 = np.any(a_np > 3) 
r2 = np.all((a_np > 3)) 
r3 = a_np >= b_np
r4 = a_np[a_np >= b_np]
r5 = np.where(a_np > 2)
r6 = np.where(a_np > 2, a_np, 0)
r7 = np.where(a_np > 2, a_np*2, 1)
r8 = np.argmax(a_np)
r9 = np.argmax(a_np, axis=1)
r10 = np.argmax(a_np, axis=0)

num = 1
r_list = [r1,r2,r3,r4,r5,r6,r7,r8,r9,r10]
r_real_list = [
    'r1 = np.any(a_np > 3)',
    'r2 = np.all((a_np > 3))', 
    'r3 = a_np >= b_np',
    'r4 = a_np[a_np >= b_np]',
    'r5 = np.where(a_np > 2)',
    'r6 = np.where(a_np > 2, a_np, 0)',
    'r7 = np.where(a_np > 2, a_np*2, 1)',
    'r8 = np.argmax(a_np)',
    'r9 = np.argmax(a_np, axis=1)',
    'r10 = np.argmax(a_np, axis=0)',
    ]
r_explain_list = [
    'np_any()항수는 괄호 안의 조건에 해당하는 요소가 있는지 판정하는 함수이다.', #r1's explain
    'np_all()함수는 괄호 안의 조건에 모든 요소가 해당하는지 판정하는 함수이다. ', #r2's explain
    '>=(비교) 연산자로 각각의 배열의 위치가 같은 요소끼리 대소비교가 가능하다.', #r3's explain
    'a_np[a_np >= b_np]로 a_np의 요소 중 []괄호 안의 조건에 맞는 요소만 리스트로 가지고 온다.', #r4's explain
    'np_where()로 괄호 안의 조건에 맞는 요소의 인덱스를 행 리스트, 열 리스트 순서로 출력한다.', #r5's explain
    'np.where(a_np > 2, a_np, 0)으로 괄호 안에 조건문에 참 요소는 a_np, 거짓 요소는 0으로 할당한다.', #r6's explain
    'np.where(a_np > 2, a_np*2, 1)으로 괄호 안에 조건문에 참 요소는 a_np*2, 거짓 요소는 1로 할당한다.', #r7's explain
    'np.argmax()함수는 괄호 안의 배열의 최대 값의 1차원 리스트로 볼 때의 인덱스를 반환한다.', #r8's explain
    'np.argmax(a_np, axis=1)함수는 a_np 배열의 최대 값의 인덱스를 행 기준으로 비교해 반환한다.', #r9's explain
    'np.argmax(a_np, axis=0)함수는 a_np 배열의 최대 값의 인덱스를 열 기준으로 비교해 반환한다.', #r10's explain
    ]
for i in range(0,len(r_list)):
  print('r{} = {}'.format(num,r_list[i]))
  print(r_real_list[i])
  print('r{}에서 {}'.format(num,r_explain_list[i]))
  print('-'*40)
  num+=1

while(True):
  exit=input('Enter를 누르면 종료\n')
  if(exit==''):break
```
